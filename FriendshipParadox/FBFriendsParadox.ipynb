The friendship paradox is the phenomenon first observed by the sociologist Scott L. Feld in 1991 that most people have fewer friends than their friends have, on average. 

Explanation: â€¦form of sampling bias in which people with greater numbers of friends have an increased likelihood of being observed among one's own friends. (Wikipedia)

The data was obtained by manually scraping (i.e. highlighting and copying) from peoples Facebook friends list.  The data was then cleaned by hand and converted to .csv format.  This method and the original code (which has since been modified) were found at:
http://controlfd.com/2016/05/18/facebook-friendship-paradox-with-how-to-guide.html

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# filename should be a .csv file in string form
#   function  (string    number  string)
def friendhist(filename, friends, name):
    ser = pd.read_table(filename)
    average = ser.mean()
    median = ser.median()
    me = friends

    # Creates histogram
    ser.hist(bins=20, color='c')
    plt.title("Histogram Friend Counts - " + name, fontsize=16)

    # Creates legend 
    red_patch = mpatches.Patch(color='r', label='Friends: %d ' % me)
    blue_patch = mpatches.Patch(color='b', label='Median: %d ' % median)
    yellow_patch = mpatches.Patch(color='y', label='Average: %d ' % average)
    plt.legend(handles=[red_patch, blue_patch, yellow_patch])

    # Plots lines at avg, median, and me
    for (x,c) in zip([average, median, me],['y','b','r']):
        plt.vlines(x,0,120,colors=c, linestyles='solid')

    plt.show()
```

```python
friendhist('myfriendnums.csv', 638, 'Ethan')
```
